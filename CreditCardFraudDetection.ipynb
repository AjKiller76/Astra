```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('fraudTrain.csv')
df1=pd.read_csv('fraudTest.csv')
```

```python
df.head(10)
```

```python
df1.head(10)
```

```python
df["trans_year"]=pd.Series(pd.to_datetime(df.trans_date_trans_time)).dt.year
df["dob"]=pd.Series(pd.to_datetime(df.dob)).dt.year
```

```python
df1["trans_year"]=pd.Series(pd.to_datetime(df1.trans_date_trans_time)).dt.year
df1["dob"]=pd.Series(pd.to_datetime(df1.dob)).dt.year
```

```python
age=pd.Series(df.trans_year-df.dob)
df["age"]=age
```

```python
age=pd.Series(df1.trans_year-df1.dob)
df1["age"]=age
```

```python
df1.head(10)
```

```python
bins=[10,18,35,60,100]
labels=['10-18','19-35','36-60','60+']
df['age_group']=pd.cut(df['age'],bins=bins,labels=labels,right=True)
```

```python
bins=[10,18,35,60,100]
labels=['10-18','19-35','36-60','60+']
df1['age_group']=pd.cut(df1['age'],bins=bins,labels=labels,right=True)
```

```python
df.head(10)
```

```python
df.isna().sum()
```

```python
df1.isna().sum()
```

```python
df.drop(axis=1,columns=["age","dob","trans_year","trans_date_trans_time"],inplace=True)
df1.drop(axis=1,columns=["age","dob","trans_year","trans_date_trans_time"],inplace=True)
```

```python
x_train=df.drop(axis=1,columns=["is_fraud"])
y_train=df["is_fraud"]
```

```python
x_test=df1.drop(axis=1,columns=["is_fraud"])
y_test=df1["is_fraud"]
```

```python
x_train.head(10)
```

```python
x_train.drop(axis=1,columns=["Unnamed: 0", "cc_num", "zip", "lat", "long", "city_pop", "unix_time", "merch_lat", "merch_long", "trans_num", "street", "first" ,"last" ,"city" ,"merchant"],inplace=True)
```

```python
x_train.head(10)
```

```python
x_train.drop(axis=1,columns=["job"],inplace=True)
```

```python
x_train.head(10)
```

```python
x_enc=pd.get_dummies(x_train,columns=["category","gender","state","age_group"],drop_first=True)
```

```python
x_test.drop(axis=1,columns=["Unnamed: 0", "cc_num", "zip", "lat", "long", "city_pop", "unix_time", "merch_lat", "merch_long", "trans_num", "street", "first" ,"last" ,"city" ,"job","merchant"],inplace=True)
```

```python
xt_enc=pd.get_dummies(x_test,columns=["category","gender","state","age_group"],drop_first=True)
```

```python
x_enc
```

```python
x_enc.columns
```

```python
xt_enc.columns
```

```python
from sklearn.linear_model import LogisticRegression
lc=LogisticRegression(penalty="l2",fit_intercept=False,random_state=7)
```

```python
lc.fit(x_enc,y_train)
```

```python
training=lc.feature_names_in_
xt_enc=xt_enc.reindex(columns=training,fill_value=0)
pred=lc.predict(xt_enc)
```

```python
from sklearn.metrics import accuracy_score
accuracy=accuracy_score(y_test,pred)
accuracy
```
